```r
set.seed(110)
sims = 1000
```

```r
sun = rep(0, sims)
work = rep(0, sims)
```

```r
for(i in 1:sims){
    weather = runif(1)
    go  = runif(1)
    if(weather <= .6){
        sun[i] = 1
        if(go <= .95){
            work[i] = 1
        }
    }
    else{
        if(go <= .3){
            work[i] = 1
        }
    }
}

mean(sun); mean(work)
```

```r
set.seed(110)
sims = 1000

Sam = rep(0, sims)
Frodo = rep(0, sims)

for(i in 1:sims){
    Sam.flip = runif(1)
    Frodo.flip = runif(1)
    if(Sam.flip <= .8){
        Sam[i] = 1
        if(Frodo.flip <= .9){
            Frodo[i] = 1
        }
    }else{
        if(Frodo.flip <= .1){
            Frodo[i] = 1
        }
    }
}
mean(Sam)
```

```r
mean(Sam[Frodo == 1])
```

```r
set.seed(110)
sims = 1000

n = 2:10

sol.a = rep(NA, length(n))
sol.e = rep(NA, length(n))

for(j in 1:length(n)){
    k = 1:n[j]
    sol.a[j] = sum((1/factorial(k))*(-1)^(k+1))

    match = rep(0, sims)

    for(i in 1:sims){
        babies = sample(1:n[j])
        ratios = babies / (1:n[j])
        if(length(ratios[ratios == 1]) > 0){
            match[i] = 1
        }
    }
    sol.e[j] = mean(match)
}
```

```r
plot(n, sol.a, main = "Hospital Matching Problem",
xlab = "n", ylab = "P(At Least One Match)",
type = "p", lwd = 3, col = "firebrick3",
ylim = c(0, 1), pch = 16)
lines(n, sol.e, col = "black", lwd = 3, type = "p", pch = 16)

abline(h = 1 - 1/exp(1))

legend("topright", legend = c("Empirical Result", "Analytical Result"),
lty = c(2, 2), lwd = c(2.5, 2.5),
col = c("black", "red"))
```

```r
set.seed(110)
sims = 1000
H = rep(0, sims)
E = rep(0, sims)
O = rep(0, sims)

for(i in 1:sims){
    flip = runif(1)
    roll = sample(1:6, 1)
    if(flip <= 1/2){
        H[i] = 1
    }
    if(roll%%2 == 0){
        E[i] = 1
    }
    if(roll%%2 == 1){
        O[i] = 1
    }
}
mean(E); mean(O); mean(H)
```

```r
mean(H[O == 1]); mean(H[E == 1])
```

```r
mean(O[H == 1]); mean(O[H == 1])
```

```r
mean(E[O == 1])
```

```r
set.seed(110)
sims = 1000

match.prob <- function(n){
    v = 365:(365 - n + 1)
    return(1 - prod(v)/365^n)
}

people = 10:50

p.match = rep(NA, length(people))

for (j in 1:length(people)){
    matches = rep(0, sims)
    for(i in 1:sims){
        bdays = sample(1:365, people[j], replace = TRUE)
        if(length(unique(bdays)) < people[j]){
            matches[i] = 1
        }
    }
    p.match[j] = mean(matches)
}

plot(1:100, sapply(1:100, function(x) match.prob(x)),
main = "P(at least one match) vs. n", 
xlab = "n", ylab = "P(at least one match)",
col = "firebrick3", type = "l",
lwd = 4)
abline(h = 1/2, col = "black")

lines(people, p.match, col = "dodgerblue4",
lwd = 4, type = "p", pch = 16)

legend("bottomright", legend = c("Empirical Result", "Analytical Result"),
lty = c(2, 1), lwd = c(2.5, 2.5),
col = c("dodgerblue4", "firebrick3"))
```

```r
set.seed(110)
sims = 1000

monty = rep(NA, sims)
car = rep(NA, sims)

for(i in 1:sims){
    doors = sample(c(1, 0, 0))

    car[i] = which(doors == 1)

    if(car[i] == 1){
        monty[i] = sample(c(2, 3), 1)
    }
    else{
        doors[1] = 1
        monty[i] = which(doors == 0)
    }
}
length(car[car == 1 & monty == 2]) / length(car[monty == 2])
```

```r
length(car[car == 3 & monty == 2]) / length(car[monty == 2])
```

```r
set.seed(110)
sims = 1000

monty = rep(NA, sims)
car = rep(NA, sims)

for(i in 1:sims){
    doors = sample(c(1, 0, 0))
    car[i] = which(doors == 1)

    player_pick = 1

    if(car[i] == player_pick){
        monty[i] = sample(setdiff(1:3, player_pick), 1)
    }
    else{
        monty[i] = setdiff(1:3, c(player_pick, car[i]))
    }
}

length(car[car == 1 & monty == 2]) / length(car[monty == 2])
```

```r
length(car[car == 3 & monty == 2]) / length(car[monty == 2])
```
